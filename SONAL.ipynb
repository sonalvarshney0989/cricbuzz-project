```python
#Q5 Calculate how many matches each team has won. Show team name and total number of wins. Display teams with the most wins first
```

```python
import requests
from collections import defaultdict

url = "https://cricbuzz-cricket.p.rapidapi.com/matches/v1/recent"

headers = {
	"x-rapidapi-key": "dc4a560bccmshb68ed9cdb6388e1p16c2eajsn9e9a2184a72c",
	"x-rapidapi-host": "cricbuzz-cricket.p.rapidapi.com"
}

response = requests.get(url, headers=headers)



data = response.json()

# Dictionary to store win counts
team_wins = defaultdict(int)

# Traverse the matches and count wins
print("typeMatches:", data.get("typeMatches"))  # See if there is any data

for match_type in data.get("typeMatches", []):
    for series in match_type.get("seriesMatches", []):
        wrapper = series.get("seriesAdWrapper", {})
        for match in wrapper.get("matches", []):
            info = match.get("matchInfo", {})
            print("Match status:", info.get("status"))  # Print every status
            # ...rest of your code...
            # Check if match result is available
            if "status" in info and "won" in info["status"].lower():
                # Try to extract the winning team name from the status string
                status = info["status"].lower()
                team1 = info.get("team1", {}).get("teamName", "N/A")
                team2 = info.get("team2", {}).get("teamName", "N/A")
                # Heuristic: the winning team is usually mentioned at the start of the status
                if team1.lower() in status:
                    team_wins[team1] += 1
                elif team2.lower() in status:
                    team_wins[team2] += 1
                else:
                    # Try to extract team name from status string
                    for t in [team1, team2]:
                        if t.lower() in status:
                            team_wins[t] += 1
                            break

# Sort teams by number of wins (descending)
sorted_teams = sorted(team_wins.items(), key=lambda x: x[1], reverse=True)

print("Team Name | Total Wins")
print("----------------------")
for team, wins in sorted_teams:
    print(f"{team} | {wins}")
```
